```python
import numpy as np
D=np.array([27,25,24,26,24])
E=np.array([26,27,27,28,27])
F=np.array(["mfcs","c++","ds","coa","mefa"])
plt.plot(F,D,color="red",label='sai k')
plt.plot(F,E,color="black",label='loki',linestyle=":",marker='^')
plt.title("marks vs subjects")
plt.legend()
plt.show()
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,27,400])
a=np.diff(A)
print("profits of company A are",a)
b=np.diff(B)
print("profits of company B are",b)
C=(["2019-2020","2020-2021","2021-2022","2022-2023"])

plt.subplot(2,1,1)
plt.title("bar graph of profit")
plt.bar(C,a,color="Black",label="profits of A")
plt.legend(loc="best")
plt.subplot(2,1,2)
plt.bar(C,b,color="green",label="profits of B")
plt.legend(loc="best")
```
